```python
(df.loc[1,'Name'][0:start] + df.loc[1,'Name'][end +1 :]).rstrip()
```

```python
def replace_substr(Name):
    start = Name.find("(")
    end = Name.find(")")
    return (Name[0:start] + Name[end + 1:]).strip()
```

```python
df['Name_trim'] = df['Name'].apply(replace_substr)
```
